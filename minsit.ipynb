```python
import torch.nn as nn
import numpy as np
import os
import torch
import torch.optim as optim
import torch.nn.functional as F
import pandas as pd
```

```python
class MLPNet(nn.Module):
    def __init__(self):
        super(MLPNet,self).__init__()
        self.fc1 = nn.Linear(784,444)
        self.fc2 = nn.Linear(444,512)
        self.fc3 = nn.Linear(512,512)
        self.fc4 = nn.Linear(512,10)
    def forward(self,x):
        x = self.fc1(x).sigmoid()
        x = self.fc2(x).sigmoid()
        x = self.fc3(x).sigmoid()
        x = self.fc4(x).sigmoid()
        return x
mlp = MLPNet()
mlp = mlp.cuda()
```

```python
os.chdir("/content/drive/MyDrive")
raw_df = pd.read_csv("mnist_test.csv")
label = raw_df["label"].values
raw_df = raw_df.drop(["label"],axis = 1)
feature = raw_df.values
n = len(feature)

test_label = torch.tensor(label[:int(n*0.1)]).to(torch.float32).cuda()
test_feature = torch.tensor(feature[:int(n*0.1)]).to(torch.float32).cuda()

train_label = torch.tensor(label[int(n*0.1):]).cuda()
train_feature = torch.tensor(feature[int(n*0.1):]).to(torch.float32).cuda()
```

```python
optimizer = optim.Adam(mlp.parameters(),lr = 0.0001)
criterion = nn.CrossEntropyLoss()
count = 0
correct = 0
num_epochs = 100
for epoch in range(num_epochs):
    for i,row in enumerate(train_feature):
        count += 1
        optimizer.zero_grad()
        outputs = mlp(row)
        loss = criterion(outputs, train_label[i])
        res = torch.argmax(outputs)
        if res == train_label[i]:
            correct += 1
        acc = float(correct/count)
        loss.backward()
        optimizer.step()
    print('epoch = %6d, loss = %12.8f, accuracy = %12.8f' % (epoch, loss,acc))
```
